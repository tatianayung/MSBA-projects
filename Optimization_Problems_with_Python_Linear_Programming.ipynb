# 1. Decision Variables

```
import pulp

D = pulp.LpVariable('D', lowBound=0) #'D' must greater than or equal to zero
S = pulp.LpVariable('S', lowBound=0) #'S' must greater than or equal to zero
```

# 2. Objective/Goal

```
profit = pulp.LpProblem('Maximum Profit', pulp.LpMaximize) # want to find maximum value
profit+=10*S+9*D # objective function formula
```

# 3. Constraints

```
"""ADD CONSTRAINT"""
profit+=7/10*S+D<=630, 'Constraint for Cutting & Dyeing'
profit+=1/2*S+5/6*D<=600, 'Constraint for Sewing'
profit+=S+2/3*D<=708, 'Constraint for Finishing'
profit+=1/10*S+1/4*D<=135, 'Constraint for I&P'

profit.solve()

print('Decision Variable Values:')

for variable in profit.variables():
  print(variable.name, '=', variable.varValue)

print('\nMaximum Profit:')
pulp.value(profit.objective)
```
